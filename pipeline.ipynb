```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")

res=classifier("I've been waiting for my java course")

print(res)
```

```python
detect = pipeline("text-generation",model='distilgpt2')

res = detect(
    "In this course i will learn,how much benefit i get",
    max_length = 30,
    num_return_sequences= 2
)
print(res)
```

```python
classifier = pipeline("zero-shot-classification")

res = classifier("This is  a course to learn about ecommerce,i will be became",
                 candidate_labels = ["educator","buisness_man","politics"]
          )
print(res)
```

```python
classifier=pipeline("text-generation")

res = classifier("what is the quality of pvc pipe")

print(res)
```
